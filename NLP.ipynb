```python
import tensorflow
from tensorflow.keras.layers.experimental.preprocessing import TextVectorization
```

```python
text_vectorizer = TextVectorization(max_tokens=None,
                                       standardize="lower_and_strip_punctuation",
                                       split="whitespace",
                                       ngrams= None,
                                       output_mode="int",
                                       output_sequence_length=None,
                                       pad_to_max_tokens=False)
```

```python
max_vocab_length = 10000
max_length = 15

text_vectorizer = TextVectorization(max_tokens = max_vocab_length,
                                    output_mode = "int",
                                    output_sequence_length=max_length)
```

```python
text_vectorizer.adapt(train_sentences)
```

```python
sample = "There is a flood in my backyard"
text_vectorizer([sample])
```

```python
random_sentence = random.choice(train_sentences)
print(f"og text:{random_sentence}")
print("Vectorised:", text_vectorizer([random_sentence]))
```

```python
winvocab = text_vectorizer.get_vocabulary()
top5words = winvocab[:5]
bottom5words = winvocab[-5:]
print(f"No. of words {len(winvocab)}")
print(f"5 Most Common Words {top5words}")
print(f"5 Most Uncommon Words {bottom5words}")
```

```python
#creating embeddings with embedding layers
```

```python
from tensorflow.keras import layers

embedding = layers.Embedding(input_dim = max_vocab_length,
                             output_dim = 128,
                             embeddings_initializer="uniform",
                             input_length = max_length)
```

```python
embedding
```

```python
random_sentence = random.choice(train_sentences)
print(f"OG Text: {random_sentence}")
sample =  embedding(text_vectorizer([random_sentence]))
print("Embeded: ",sample)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline

model_0 = Pipeline([
    ("tfidf", TfidfVectorizer()),
    ("clf", MultinomialNB())

])

model_0.fit(train_sentences, train_labels)
```

```python
baseline_score = model_0.score(val_sentences, val_labels)
print(f"Accuracy {baseline_score*100:.2f}%")
```

```python
baseline_preds = model_0.predict(val_sentences)
baseline_preds[:20]
```
